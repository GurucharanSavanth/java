## Code Modification with bias
* OR([1 1]) = 1
* AND([0 1]) = 0
* XOR([1 0]) = 1
* XNOR([0 0]) = 1

```python
import numpy as np

# Define Unit Step Function
def unitStep(v):
   if v >= 0:
       return 1
   else:
       return 0

# Design Perceptron Model
def perceptronModel(x, w, b):
   v = np.dot(w, x) + b
   y = unitStep(v)
   return y

# Define functions
def OR_logicFunction(x):
   weights = [1, 1]
   bias = -0.5
   return perceptronModel(x, weights, bias)

def AND_logicFunction(x):
   weights = [1, 1]
   bias = -1.5
   return perceptronModel(x, weights, bias)

def XOR_logicFunction(x):
   weights = [1, 1]
   bias = -1
   return perceptronModel(x, weights, bias)

def XNOR_logicFunction(x):
   weights = [-1, -1]
   bias = 0.5
   return perceptronModel(x, weights, bias)

# Test cases
test1 = np.array([0, 0])
test2 = np.array([0, 1])
test3 = np.array([1, 0])
test4 = np.array([0, 0])

# Execute the code for each operation

print("")
print("OR({}) = {}".format(test1, OR_logicFunction(test1)))
print("AND({}) = {}".format(test2, AND_logicFunction(test2)))
print("XOR({}) = {}".format(test3, XOR_logicFunction(test3)))
print("XNOR({}) = {}".format(test4, XNOR_logicFunction(test4)))
```

## Code Modification without bias

---
* OR([1 1]) = 1
* AND([0 1]) = 0
* XOR([1 0]) = 1
* XNOR([0 0]) = 1

```python
import numpy as np

# Define Unit Step Function
def unitStep(v):
    if v >= 0:
        return 1
    else:
        return 0

# Design Perceptron Model with Bias
def perceptronModelWithBias(x, w, b):
    v = np.dot(w, x) + b
    y = unitStep(v)
    return y

# Design Perceptron Model without Bias
def perceptronModelWithoutBias(x, w):
    v = np.dot(w, x)
    y = unitStep(v)
    return y

# Define functions for different logic operations without bias
def OR_logicFunctionWithoutBias(x):
    weights = [1, 1]
    return perceptronModelWithoutBias(x, weights)

def AND_logicFunctionWithoutBias(x):
    weights = [1, -1]
    return perceptronModelWithoutBias(x, weights)

def XOR_logicFunctionWithoutBias(x):
    weights = [1, -1]
    return perceptronModelWithoutBias(x, weights)

def XNOR_logicFunctionWithoutBias(x):
    weights = [-1, 1]
    return perceptronModelWithoutBias(x, weights)

# Test cases
test1 = np.array([1, 1])
test2 = np.array([0, 1])
test3 = np.array([1, 0])
test4 = np.array([0, 0])

print("\n Code Without Bias:")
print("OR({}) = {}".format(test1, OR_logicFunctionWithoutBias(test1)))
print("AND({}) = {}".format(test2, AND_logicFunctionWithoutBias(test2)))
print("XOR({}) = {}".format(test3, XOR_logicFunctionWithoutBias(test3)))
print("XNOR({}) = {}".format(test4, XNOR_logicFunctionWithoutBias(test4)))
```
